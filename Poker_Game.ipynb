```python
import random
 
suits=("Hearts","Diamonds","Spades","Clubs")
ranks=('Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King','Ace')
values={'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,'Jack':10,'Queen':10,'King':10,'Ace':10}
    
playing=True
```

```python
class Card():
     
        def __init__(self,suit,rank):
            self.suit=suit
            self.rank=rank
            
        def __str__(self):
            return self.rank+" of " +self.suit
```

```python
class Deck():
    def __init__(self):
        self.list=[]
        for suit in suits:
            for rank in ranks:
                self.list.append(Card(suit,rank))
                
    def __str__(self):
        deck_comp=" "
        for card in self.list:
            deck_comp+= "\n"+ card.__str__()
        return "The Deck has:" +deck_comp
    
    def shuffle(self):
        random.shuffle(self.list)
            
    def deal(self):
        single_card=self.list.pop()
        return single_card
    
```

```python
class Hand():
    
    def __init__(self):
        self.list1=[]
        self.value=0
        self.aces=0
        
    def add_card(self,card):
        self.list1.append(card)
        self.value+=values[card.rank]
        
        if card.rank == "Ace":
            self.aces+=1
         
        
    def adjust_ace(self):
        
        while self.value>21 and self.aces:
            self.value-=10
            self.aces-=1
        
        
        
```

```python
class Chip():
    def __init__(self,total=100):
        self.total=total
        self.bet=0
        
    def win_bet(self):
        self.total+=self.bet
        
    def lose_bet(self):
        self.total-=self.bet
```

```python
def take_bet(chips):
    
    while True:
        try:
            chips.bet=int(input("Please Enter the no of chips you want"))
        except:
            print("please enter a valid no")
        else:
            if chips.bet>chips.total:
                print(f"You dont't have enough chips .Your chips are left are:{chips.total}")
            else:
                break
```

```python
def hit(deck,hand):
    
        
    single_card=deck.deal()
    hand.add_card(single_card)
    hand.adjust_ace()
```

```python
def hit_or_stand(deck,hand):
    global playing
    
    while True:
         
            x=input("Want to Hit or Stand? Enter h or s")
            
            if(x[0].lower() == 'h'):
                hit(deck,hand)
                
            elif (x[0].lower() == 's'):
                print("As you want to stand it's dealer turn")
                playing = False
                
            else:
                print("I don't understand what you want to say .Plaese enter h or s")
                continue
                
            break
```

```python
def player_burst(player,dealer,chips):
    print("Player burst")
    chips.lose_bet()
    
def player_win(player,dealer,chips):
    print("Player Wins")
    chips.win_bet()
    
def dealer_burst(player,dealer,chips):
    print("Player Wins , Dealer burst")
    chips.win_bet()
    
def dealer_win(player,dealer,chips):
    print("Dealer win ,Player burst ")
    chips.lose_bet()
    
def push(player,dealer):
    print("Thera's a tie between Player and Dealer")
    
```

```python
def show_some(player,dealer):
    
    print("Dealer's cards:")
    print("One card of dealer is hidden")
    print(dealer.list1[1])
    print("Player's cards:")
    for cards in player.list1:
        print(cards)
        
def show_all(player,dealer):
    print("Dealer's cards:")
    for card in dealer.list1:
        print(card)
    print("Player's cards:")
    for cards in player.list1:
        print(cards)
```
