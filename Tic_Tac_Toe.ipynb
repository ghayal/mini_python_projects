```python
def player_input():
    marker =''
        
    while(marker !='O' and marker !='X'):
        marker=input("Player 1, choose 'X' OR 'O'")
        
    Player1= marker
    
    if Player1 == 'X':
        Player2='O'
    else:
        Player2='X'
    return (Player1,Player2)
```

```python
import random

def choice():
    
    flip = random.randint(0,1) 
    if flip == 0:
        return 'Player_1'
    else:
        return 'Player_2'
    
```

```python
def space_free(board,position):
    
    m=int(position)
        
    return board[position] == ' '
```

```python
def Full_board(board):
    for m in range(1,10):
        if space_free(board, m):
            return False
        
    return True
```

```python
def player_choice(board):
    
    position = 0
    while position not in range (1,10) or not space_free(board,position):
        position =int(input("choose the no between :1-9"))
               
        
    return position
```

```python
#START THE GAME

print("Welcome to Tic tac toe!")

while True:
    #play the game 
    #set board , marker , whose gona first:
    board=[' ']*10
    Player1_marker, Player2_marker= player_input() 
    
    turn=choice()
    print(turn +" will go first")
    #game play
    
    play_game=input("Ready to play? y or n")
    
    if play_game == 'y':
        game_on=True
    else:
        game_on=False
        
    #start the game
    while game_on:
         #player 1 turn
        if turn == 'Player_1':
            
            #see display board
            display_board(board)
            #choose the position
            position=player_choice(board)            
            #marker is placed 
            place_marker(board,Player1_marker,position)
            #display_board(board)
            #whether win or not
            if win(board,Player1_marker):
                display_board(board)
                print("Player 1 has won!")
                game_on =False
            else:
                if Full_board(board):
                    display_board(board)
                    print("TIE !!")
                    game_on =False
                else:
                    turn ='Player_2'
            
        else:
            #see display board
            display_board(board)
            #choose the position
            position=player_choice(board)
            #marker is placed 
            place_marker(board,Player2_marker,position)
            #whether win or not
            if win(board,Player1_marker):
                display_board(board)
                print("Player 2 has won!")
                game_on =False
            else:
                if Full_board(board):
                    display_board(board)
                    print("TIE !!")
                    game_on =False
                else:
                    turn ='Player_1'
                    
                    
            #player 2 turn
        
    if not replay():
        break
```
